```python
import sqlite3
import pandas as pd
import datetime
from dateutil.relativedelta import *
import random
```

# Чтение новых данных из файла

```python
cred_types = ['РК', 'ВКЛ', 'НКЛ', 'ОВ','Г']
start_date = datetime.datetime.today()

generated_data = list()

for i in range(50):
    for cred_type in cred_types:
        one_row = list()
        one_row.append(start_date.date())
        one_row.append(cred_type)
        one_row.append(random.randint(1,200))
        one_row.append(random.randint(1,20000))
        one_row.append(random.randint(1,200))
        one_row.append(0)
        one_row.append((start_date+relativedelta(months=-i)).date())
        
        generated_data.append(one_row)

new_data = pd.DataFrame(generated_data, columns = ['load_date', 'product_type', 
                                       'in_default', 'healthy', 'will_default', 'deleted', 'report_date'])
new_data.head()
```

```python
database_path = '../chartsite/db.sqlite3'
conn = sqlite3.connect(database_path)
```

```python
new_data.to_sql('dashboard_report_counts', conn, index_label='id', if_exists='replace')
```
